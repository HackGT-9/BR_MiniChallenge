```python
import pandas as pd 
import numpy as np
import requests 
import urllib.request, json
import matplotlib.pyplot as plt
```

```python
url = 'https://mocki.io/v1/43d2c1e0-1428-4c2b-9ae9-05513371693b'
response = requests.get(url)
data = response.json()
```

```python
data = json.loads(response.text)
res = {}
for each in data['stocks']:
    for dude in each:
        if dude == 'stockSymbol':
            res[each[dude]] = []
        else:
            break
for man in data['stocks']:
    for each in man['prices']:
        res[man['stockSymbol']].append(each['price'])
res
```

```python
print("AMZN OR MSFT OR GOOGL?")
stock = input() 
def max_prof(res, stck):
    price = res[stck]
    length = len(price)
    print(price)
    max_diff = price[1] - price[0]
    min_elem = price[0]
    for ind in range(1, length):
        if (price[ind] - min_elem > max_diff):
            max_diff = price[ind] - min_elem
        if (price[ind] < min_elem):
            min_elem = price[ind]
    return max_diff
max_prof(res, stock)
```

```python
def plot(stck):
    rates = res[stck]
    days = list(range(0, len(rates)))
    print(stck)
    plt.bar(days, rates, color='Blue')
    plt.plot(rates, color = "Green")
    plt.show()
for samurai in res:
    plot(samurai)
```
